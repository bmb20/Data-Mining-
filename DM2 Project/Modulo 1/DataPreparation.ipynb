CORRELAZIONI TRAIN: frame_count e length (per tutte le divisioni) --> decidere quale delle due eliminare (DM1 eliminato frame_count)

```python
#train_corr = df_TRAIN.corr(method="pearson").style.background_gradient(cmap='coolwarm', vmin=-1, vmax=1)
```

```python
train_corr = df_TRAIN.corr()
```

```python
train_corr
```

```python
correlated_features = set()
for i in range(len(train_corr.columns)):
    for j in range(i):
        if abs(train_corr.iloc[i, j]) > 0.9:
            colname = train_corr.columns[i]
            print(colname)
            correlated_features.add(colname)
```

```python
correlated_features.remove('length_w4')
correlated_features.remove('length_w3')
correlated_features.remove('length_w2')
correlated_features.remove('length_w1')
```

```python
len(correlated_features)
```

```python
correlated_features.add('frame_count')
```

```python
len(correlated_features)
```

```python
print(correlated_features)
```

```python
#eliminare correlated_features
df_TRAIN.drop(labels=correlated_features, axis=1, inplace=True)
```
